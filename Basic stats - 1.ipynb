## Handle Missing Values

```python
print(df.isnull().sum())
```

# Identify Numerical & Categorical Columns

```python
# Numerical columns
num_cols = df.select_dtypes(include=['int', 'float']).columns
print("Numerical Columns:", num_cols)

# Categorical columns
cat_cols = df.select_dtypes(include=['object']).columns
print("Categorical Columns:", cat_cols)
```

# Descriptive Analytics (Basic Statistics)

```python
# Mean
print(df[num_cols].mean())

# Median
print(df[num_cols].median())

# Mode
print(df[num_cols].mode().iloc[0])

# Standard Deviation
print(df[num_cols].std())
```

# Brief Interpretation Of These Statistics.

```python
# Mean → Represents the average of all values in the dataset, calculated by dividing the total sum by the number of observations, and is sensitive to extreme values (outliers).

# Median → Represents the middle value of the dataset when arranged in order, dividing it into two equal halves, and is not affected by outliers, making it suitable for skewed data.

# Mode → Represents the most frequently occurring value in the dataset, useful for identifying common values or categories, and a dataset may have one, multiple, or no modes.

# Standard Deviation → Measures the spread or variability of data around the mean, where a higher value indicates greater dispersion and a lower value indicates that data points are closer to the mean.
```

# Data Visualization

## Histograms

```python
df[num_cols].hist(figsize=(10,8))
plt.show()
```

## Box Plots

```python
for col in num_cols:
    plt.figure()
    sns.boxplot(x=df[col])
    plt.title(col)
    plt.show()
```

## Bar Chart (Categorical Columns)

```python
for col in cat_cols:
    plt.figure()
    df[col].value_counts().plot(kind='bar')
    plt.title(col)
    plt.show()
```
