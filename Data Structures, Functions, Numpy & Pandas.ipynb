1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
# Creating list of 5 integers
numbers = [10, 20, 30, 40, 50]
print(numbers)
```

```python
numbers.append(60) #append means add to item to end of the list
print("append:", numbers)
```

```python
numbers.extend([70, 80]) #add multiple items to end of the list
print("extend:", numbers)
```

```python
numbers.insert(2, 25) # inserts element at a specific position
print("insert:", numbers)
```

```python
numbers.remove(40) #removes the item at value based.
print("remove:", numbers)
```

```python
numbers.pop(3) #remove the item using index based
print("pop:", numbers)
```

```python
numbers.clear() #clears all items in the list
print("clear:", numbers)
```

```python
print("Index of 20:", numbers.index(20)) #returns index value
```

```python
print("Count of 20:", numbers.count(20)) #it counts the number of occurances
```

```python
 numbers.sort() #sort the llist in asscending order
print("sort:", numbers)
```

```python
numbers.reverse() #reverse the list
print("reverse:", numbers)
```

7.	Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
# Given number
num = 8

# Check conditions
if num > 0:
    print("The number is positive.")
    
    if num % 2 == 0:
        print("The number is also even.")
    else:
        print("The number is odd.")

elif num == 0:
    print("The number is zero.")

else:
    print("The number is negative.")
```

11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
# Creating DataFrame with 10 rows
data = {
    "Name": ["uday", "Anita", "Ravi", "Priya", "Kiran", "Sneha", "Arjun", "Meena", "Ravi", "Priya"],
    "Age": [22, 23, 21, 24, 22, 23, 25, 21, 21, 24],
    "City": ["Delhi", "Mumbai", "Chennai", "Bangalore", "Hyderabad", "Pune", "Delhi", "Mumbai", "Chennai", "Bangalore"],
    "Salary": [50000, 60000, np.nan, 55000, 52000, np.nan, 65000, 48000, 50000, 55000]
       }

df = pd.DataFrame(data)

print("DataFrame:\n", df)
```

use info

```python
df.info()
```

use describe

```python
df.describe()
```

Select 'Name' and 'City' columns

```python
df[['Name', 'City']]
```

Drop 'City' column

```python
df = df.drop('City', axis=1)
print(df)
```

Fill any missing values in 'Salary' column with the mean

```python
df['Salary']=df['Salary'].fillna(value=df['Salary'].mean())
print(df)
```

Remove the duplicate rows

```python
df = df.drop_duplicates()
print(df)
```
