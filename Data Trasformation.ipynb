```python
#outlier capping 
def outlier_capping(df,column):
    q1=df[column].quantile(0.25)
    q3=df[column].quantile(0.75)
    iqr=q3-q1
    lower_extreme=q1-1.5*iqr
    high_extreme=q3+1.5*iqr
    df[column]=df[column].apply(lambda x: lower_extreme if x<lower_extreme else high_extreme if x>high_extreme else x)
for col in df.select_dtypes(include=['int','float']).columns:
    outlier_capping(df,col)
```

### 2. Encoding Techniques:

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
# unique values count for each categorical column

for col in categorical_cols:
    print(f"{col} : {df[col].nunique()} unique values")
```

```python
# Select Columns with Less Than 5 Categories
one_hot_cols = []

for col in categorical_cols:
    unique_count = len(df[col].unique())
    
    if unique_count < 5:
        one_hot_cols.append(col)

print("Columns for One-Hot Encoding:")
print(one_hot_cols)
```

```python
one_hot= OneHotEncoder(sparse_output=False, handle_unknown='ignore') #sparse is used for the maintain the memory,handle_unknown is used for,if column is not then it takes the columns
# Apply One-Hot Encoding
encoded_data = pd.DataFrame(data=one_hot.fit_transform(df[one_hot_cols]),columns=one_hot.get_feature_names_out(one_hot_cols))
```

```python
# Shape of one hot encoded data
print(encoded_data.shape)
```

```python
# Drop original categorical columns
df = df.drop(columns=one_hot_cols)
```

```python
# Add encoded columns to dataset
df = pd.concat([df, encoded_data], axis=1)
```

```python
# Display dataset
print(df.head())
```

```python
# Select columns for Label Encoding
label_encode_cols = []

for col in categorical_cols:
    
    # Check column exists in dataframe
    if col in df.columns:
        
        # Select columns having 5 or more unique values
        if df[col].nunique() >= 5:
            label_encode_cols.append(col)

print("Columns for Label Encoding:")
print(label_encode_cols)
```

```python
label_encoder = LabelEncoder()

for col in label_encode_cols:
    df[col] = label_encoder.fit_transform(df[col])
# Store encoded columns
encoded_df = df[label_encode_cols]
```

```python
#shape of label encoded data
print(encoded_df.shape)
```

```python
# Drop original columns
df = df.drop(columns=label_encode_cols)
```

```python
# Add encoded columns to dataset
df = pd.concat([df, encoded_df], axis=1)
```

```python
# Display dataset
print(df.head())
```

### 3. Feature Engineering:

```python
# Feature 1: Age Group Feature
```

```python
# Create Age Group Feature

df['age_group'] = pd.cut(
    df['age'],
    bins=[0, 25, 45, 65, 100],
    labels=['Young', 'Adult', 'Middle_Aged', 'Senior']
)

print(df[['age', 'age_group']].head())
```

```python
# Feature 2: Hours per Age Feature
```

```python
# Create Hours per Age Feature

df['hours_per_age'] = df['hours_per_week'] / df['age']

print(df[['hours_per_week', 'age', 'hours_per_age']].head())
```

```python
# feature 3: Capital Total Feature
```

```python
# Combine capital gain and capital loss

df['capital_total'] = df['capital_gain'] - df['capital_loss']

print(df[['capital_gain', 'capital_loss', 'capital_total']].head())
```
