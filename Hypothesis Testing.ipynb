### Confidence Intervel

```python
import numpy as np
import scipy.stats as stats
```

```python
#given data
confidence = 0.99
alpha = 1 - confidence
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])
```

```python
n=len(data)
n
```

```python
mean = np.mean(data)
mean
```

```python
#Sample standard deviation (ddof=1 for sample std)
std_dev = np.std(data, ddof=1)
std_dev
```

```python
#t-critical value
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
t_critical
```

```python
# Margin of error
margin_error = t_critical * (std_dev / np.sqrt(n))
```

```python
#Confidence interval
lower_bound = mean - margin_error
upper_bound = mean + margin_error
```

```python
print("Sample Mean:", mean)
print("Sample Standard Deviation:", std_dev)
print("t-Critical Value:", t_critical)
print("Margin of Error:", margin_error)
print("99% Confidence Interval:", (lower_bound, upper_bound))
```

```python
#given data
confidence = 0.99
alpha = 1 - confidence
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])
# Known population standard deviation
sigma = 0.2
```

```python
n=len(data)
n
```

```python
n=len(data)
n
```

```python
#Z-critical value
z_critical = stats.norm.ppf(1 - alpha/2)
```

```python
#margin of error
margin_error = z_critical * (sigma / np.sqrt(n))
```

```python
#Confidence interval
lower_bound = mean - margin_error
upper_bound = mean + margin_error
```

```python
print("Sample Mean:", mean)
print("Z-Critical Value:", z_critical)
print("Margin of Error:", margin_error)
print("99% Confidence Interval:", (lower_bound, upper_bound))
```

### Hypothesis Testing

```python
import numpy as np
from scipy.stats import t
```

```python
#given data
sample_mean=3050
theoritical_meean=1000+5*600
standard_deviation=5*25
sample_size=25
alpha=0.05
```

```python
#calculating t_statistics
standard_value=standard_deviation/np.sqrt(sample_size)
standard_value
```

```python
t_statistics=(sample_mean-theoritical_meean)/standard_value
t_statistics
```

```python
dof=sample_size-1
critical_value=t.ppf(1-alpha,dof)
```

```python
critical_value
```

```python
# Step 4: Decision
if t_statistics > critical_value:
    print("Reject H0")
else:
    print("Fail to Reject H0")
```

```python
# Step 5: Conclusion
if t_statistics > critical_value:
    print("Conclusion: Costs are higher than expected.")
else:
    print("Conclusion: No evidence that costs are higher.")
```
